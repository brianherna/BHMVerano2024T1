```python
# Datos de entrada
dt = [250, 790, 470, 510, 500, 330,
    300, 750, 580, 740, 420, 260,
    440, 770, 400, 540, 530, 330,
    390, 380, 340, 460, 530, 260,
    410, 360, 270, 510, 630, 330]
# Datos ordenados
dt = OrdenarArregloAsendente.ordenar_asc(dt)

# Obtenemos clases, lim inf, lim. sup, y marcas de clase
clases, lim_inf, lim_sup, mrks = ClasesAgrupadas.clases_grouped(dt)

# Frecuencia absoluta
fa = CalcularFABS.fa_grouped(lim_inf, lim_sup, dt)

# Frecuencia relativa
fr = FrecRelativa.frec_rel(fa)

# Frecuencia relativa acumulada
facum = FrecAcumulada.frec_acum(fr)

# Imprecion en tabla para todos los datos
TablaCompleta.tabla_grouped(clases, lim_inf, lim_sup, mrks, fa, fr, facum)
# Llamada de las funciones para graficar
GraficoHistograma.plot_grafHist(fr, mrks, clases)# Grafico Histograma
#Si mrks es una lista de marcas de clase, entonces valores_ref_eje = lista de nmeros que van desde 1 hasta el nmero de marcas de clase
valores_ref_eje = list(range(1, len(mrks) + 1)) #  Marcas de clase en el eje x 
GraficoOjiva.plot_grafOjiva(mrks, valores_ref_eje, facum) # Grafico de Ojiva
```
